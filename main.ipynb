```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**(1/2)

def taylor(x, grau):
    a = 4
    if grau == 0:
        return f(a)
    elif grau == 1:
        return f(a) + (1/2)*(a**(-1/2))*(x-a)
    elif grau == 2:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2
    elif grau == 3:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2 + (3/8)*(a**(-5/2))*(x-a)**3/6
    elif grau == 4:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2 + (3/8)*(a**(-5/2))*(x-a)**3/6 - (15/16)*(a**(-7/2))*(x-a)**4/24

x = np.linspace(0, 8)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, taylor(x, 1), label='taylor(x, 1)')
plt.plot(x, taylor(x, 2), label='taylor(x, 2)')
plt.plot(x, taylor(x, 3), label='taylor(x, 3)')
plt.plot(x, taylor(x, 4), label='taylor(x, 4)')
```
