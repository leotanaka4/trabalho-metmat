a)<br>
Polinômio de Taylor de grau $n$ de $f$ em $a$:
$$
P_{n}(x) = \sum_{n}^{i=0} \frac{f^{(n)}(a)}{i!} (x - a)^{i}
$$
Função:
$$
f(x) = \sqrt{x}
$$
Polinômio de Taylor de grau 1 de $f$ em $a$:
$$
P_{1}(x) = f(a) + \frac{1}{2} a^{\frac{-1}{2}} (x - a)
$$
Polinômio de Taylor de grau 2 de $f$ em $a$:
$$
P_{2}(x) = f(a) + \frac{1}{2} a^{\frac{-1}{2}} (x - a) - \frac{1}{4} \frac{a^{\frac{-3}{2}}(x - a)^2}{2!}
$$
Polinômio de Taylor de grau 3 de $f$ em $a$:
$$
P_{2}(x) = f(a) + \frac{1}{2} a^{\frac{-1}{2}} (x - a) - \frac{1}{4} \frac{a^{\frac{-3}{2}}(x - a)^2}{2!} + \frac{3}{8} \frac{a^{\frac{-5}{2}}(x - a)^3}{3!}
$$
Polinômio de Taylor de grau 4 de $f$ em $a$:
$$
P_{2}(x) = f(a) + \frac{1}{2} a^{\frac{-1}{2}} (x - a) - \frac{1}{4} \frac{a^{\frac{-3}{2}}(x - a)^2}{2!} + \frac{3}{8} \frac{a^{\frac{-5}{2}}(x - a)^3}{3!} - \frac{15}{16} \frac{a^{\frac{-7}{2}}(x - a)^4}{4!}
$$

```python
def f(x):
    return x**(1/2)

def taylor(x, grau):
    a = 4
    if grau == 0:
        return f(a)
    elif grau == 1:
        return f(a) + (1/2)*(a**(-1/2))*(x-a)
    elif grau == 2:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2
    elif grau == 3:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2 + (3/8)*(a**(-5/2))*(x-a)**3/6
    elif grau == 4:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2 + (3/8)*(a**(-5/2))*(x-a)**3/6 - (15/16)*(a**(-7/2))*(x-a)**4/24

x = np.linspace(0, 8, 500)
plt.rcParams["figure.figsize"]=18,12
fig, ax = plt.subplots(2, 2)

ax[0,0].plot(x,taylor(x, 1),"red", label="P1(x)")
ax[0,0].plot(x,f(x),"black", label="f(x)")
ax[0,0].set_title("Polinômio de Taylor de Grau 1")
ax[0,0].set_xlabel("x", loc="right")
ax[0,0].set_ylabel("y", loc="top")
ax[0,0].legend()

ax[0,1].plot(x,taylor(x, 2),"red", label="P2(x)")
ax[0,1].plot(x,f(x),"black", label="f(x)")
ax[0,1].set_title("Polinômio de Taylor de Grau 2")
ax[0,1].set_xlabel("x", loc="right")
ax[0,1].set_ylabel("y", loc="top")
ax[0,1].legend()

ax[1,0].plot(x,taylor(x, 3),"red", label="P3(x)")
ax[1,0].plot(x,f(x),"black", label="f(x)")
ax[1,0].set_title("Polinômio de Taylor de Grau 3")
ax[1,0].set_xlabel("x", loc="right")
ax[1,0].set_ylabel("y", loc="top")
ax[1,0].legend()

ax[1,1].plot(x,taylor(x, 4),"red", label="P4(x)")
ax[1,1].plot(x,f(x),"black", label="f(x)")
ax[1,1].set_title("Polinômio de Taylor de Grau 4")
ax[1,1].set_xlabel("x", loc="right")
ax[1,1].set_ylabel("y", loc="top")
ax[1,1].legend()
```

Os polinômios de Taylor de n-ésimo grau de $f$ em $a$ são somas parciais da Série de Taylor que representa a $f(x)$. Como $f$ é a soma de sua Série de Taylor, sabemos que $P_{n}(x) \rightarrow f(x)$ quando $n \rightarrow \infty$ e, assim, $P_{n}(x)$ pode ser usado como uma aproximação para $f: f(x)$ aproximadamente igual a $P_{n}(x)$. Então, a cada vez que o grau do Polinômio é aumentado, a $P_{n}(x)$ se aproxima da $F(x)$. Assim, vai convergindo para $f(x)$.

b)<br>

Teorema de Taylor:

Seja $f(x)$ uma função tal que $f$ e todas as suas derivadas existem em algum
intervalo aberto $I$ e seja $a$ um número qualquer de $I$. Então, para cada $x$ $\in$ $I$,
existe um número $c$ entre $a$ e $x$, tal que:
$$
f(x) = f(a) + f'(x)(x - a) + ... + \frac{f^{(n)}(x - a)^{n}}{n!} + R_{n}(x)
$$
onde Resto de ordem $n$ $(R_{n}(x))$:
$$
R_{n}(x) = \frac{f^{(n+1)}(c)(x - a)^{(n+1)}}{(n + 1)!}
$$
Resto de ordem 4 $(R_{4}(x))$:
$$
R_{4}(x) = \frac{105}{32} \frac{c^{\frac{-9}{2}}(x - a)^5}{5!}
$$

```python
def Rn(x):
    a = 4
    return (105/32)*(a**(-9/2))*(x-a)**5/120

print("A exatidão da aproximação é "+ str(Rn(4.5)))

x = np.linspace(4, 5, 500)
fig, ax = plt.subplots()
ax.plot(x, Rn(x), "red", label="Rn(x)")
ax.set_title("Desigualdade de Taylor: Rn(x)")
ax.set_xlabel("x", loc="right")
ax.set_ylabel("y", loc="top")
ax.legend()
```

Equação da resistividade:
$$
\rho(t) = \rho_{20}e^{\alpha(t - 20)}
$$
Aproximação Linear:
$$
P_{1}(t) = \rho(a) + \rho_{20}\alpha e^{\alpha(a - 20)} (t - 20)
$$
Aproximação Quadrática:
$$
P_{2}(t) = \rho(a) + \rho_{20}\alpha e^{\alpha(a - 20)} (t - 20) + \frac{\rho_{20}\alpha^{2}e^{\alpha(a - 20)} (t - 20)^{2}}{2!}
$$

```python
a = 0.0038
p20 = 1.6 * 10**(-8)

def p(t):
    return p20 * np.exp(a*(t-20))

def linear(t):
    return p(20) + p20 * a * np.exp(a*(20 - 20)) * (t-20)

def quadratica(t):
    return p(20) + p20 * a * np.exp(a*(20 - 20)) * (t-20) + p20 * a**2 * np.exp(a*(20 - 20)) * (t-20)**2 / 2
```

b)

```python
t = np.linspace(-250, 1000, 12500)

fig, ax = plt.subplots()
ax.plot(t, p(t), "black", label="Função")
ax.plot(t, linear(t), "red", label="Linear")
ax.plot(t, quadratica(t), "orange", label="Quadrática")
ax.set_title("Gráfico da função e polinômios de Taylor")
ax.set_xlabel("t", loc="right")
ax.set_ylabel("p", loc="top")
ax.legend()
```
