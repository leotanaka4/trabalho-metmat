# Questão 1

a)

```python
def f(x):
    return x**(1/2)

def taylor(x, grau):
    a = 4
    if grau == 0:
        return f(a)
    elif grau == 1:
        return f(a) + (1/2)*(a**(-1/2))*(x-a)
    elif grau == 2:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2
    elif grau == 3:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2 + (3/8)*(a**(-5/2))*(x-a)**3/6
    elif grau == 4:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2 + (3/8)*(a**(-5/2))*(x-a)**3/6 - (15/16)*(a**(-7/2))*(x-a)**4/24

x = np.linspace(0, 8)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, taylor(x, 1), label='taylor(x, 1)')
plt.plot(x, taylor(x, 2), label='taylor(x, 2)')
plt.plot(x, taylor(x, 3), label='taylor(x, 3)')
plt.plot(x, taylor(x, 4), label='taylor(x, 4)')
```

## Questão 2

a)


```python
a = 0.0038
p20 = 1.6 * 10**(-8)

def p(t):
    return p20 * np.exp(a*(t-20))

def linear(t):
    return p(20) + p20 * a * np.exp(a*(20 - 20)) * (t-20)

def quadratica(t):
    return p(20) + p20 * a * np.exp(a*(20 - 20)) * (t-20) + p20 * a**2 * np.exp(a*(20 - 20)) * (t-20)**2 / 2
```

b)

```python
t = np.linspace(-250, 1000)
plt.plot(t, p(t), label='p(t)')
plt.plot(t, linear(t), label='linear')
plt.plot(t, quadratica(t), label='quadratica')
```
