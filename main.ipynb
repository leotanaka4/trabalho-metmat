# Questão 1

a)

```python
def f(x):
    return x**(1/2)

def taylor(x, grau):
    a = 4
    if grau == 0:
        return f(a)
    elif grau == 1:
        return f(a) + (1/2)*(a**(-1/2))*(x-a)
    elif grau == 2:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2
    elif grau == 3:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2 + (3/8)*(a**(-5/2))*(x-a)**3/6
    elif grau == 4:
        return f(a) + (1/2)*(a**(-1/2))*(x-a) - (1/4)*(a**(-3/2))*(x-a)**2/2 + (3/8)*(a**(-5/2))*(x-a)**3/6 - (15/16)*(a**(-7/2))*(x-a)**4/24

x = np.linspace(0, 8, 500)
plt.rcParams["figure.figsize"]=18,9
fig, ax = plt.subplots(2, 2)

ax[0,0].plot(x,taylor(x, 1),"red")
ax[0,0].plot(x,f(x),"black")
ax[0,0].set_title("Polinômio de Taylor de Grau 1")
ax[0,0].set_xlabel("x", loc="right")
ax[0,0].set_ylabel("y", loc="top")

ax[0,1].plot(x,taylor(x, 2),"red")
ax[0,1].plot(x,f(x),"black")
ax[0,1].set_title("Polinômio de Taylor de Grau 2")
ax[0,1].set_xlabel("x", loc="right")
ax[0,1].set_ylabel("y", loc="top")

ax[1,0].plot(x,taylor(x, 3),"red")
ax[1,0].plot(x,f(x),"black")
ax[1,0].set_title("Polinômio de Taylor de Grau 3")
ax[1,0].set_xlabel("x", loc="right")
ax[1,0].set_ylabel("y", loc="top")

ax[1,1].plot(x,taylor(x, 4),"red")
ax[1,1].plot(x,f(x),"black")
ax[1,1].set_title("Polinômio de Taylor de Grau 4")
ax[1,1].set_xlabel("x", loc="right")
ax[1,1].set_ylabel("y", loc="top")
```

Os polinômios de Taylor de n-ésimo grau de f em a são somas parciais da Série de Taylor que representa a f(x). Como f é a soma de sua Série de Taylor, sabemos que Pn(x) -> f(x) quando n -> infinito e, assim, Pn(x) pode ser usado como uma aproximação para f: f(x) aproximadamente igual a Pn(x). Então, a cada vez que o grau do Polinômio é aumentado, a Pn(x) se aproxima da F(x). Assim, vai convergindo para f(x).

## Questão 2

a)


```python
a = 0.0038
p20 = 1.6 * 10**(-8)

def p(t):
    return p20 * np.exp(a*(t-20))

def linear(t):
    return p(20) + p20 * a * np.exp(a*(20 - 20)) * (t-20)

def quadratica(t):
    return p(20) + p20 * a * np.exp(a*(20 - 20)) * (t-20) + p20 * a**2 * np.exp(a*(20 - 20)) * (t-20)**2 / 2
```

b)

```python
t = np.linspace(-250, 1000, 12500)

fig, ax = plt.subplots()
ax.plot(t, p(t), "black")
ax.plot(t, linear(t), "red")
ax.plot(t, quadratica(t), "orange")
ax.set_title("Gráfico da função e polinômios de Taylor")
ax.set_xlabel("t", loc="right")
ax.set_ylabel("p", loc="top")
```

```python
x = []
y = []

for i in t:
    if (0.99 * p(i) <= linear(i) <= 1.01 * p(i)):
        x.append(i)
        y.append(linear(i))

print("O intervalo começa de %.1f °C e termina em %.1f °C" % (x[0], x[-1]))

fig, ax = plt.subplots()
ax.plot(x, y, "green")
ax.set_title("Intervalo de Valores")
ax.set_xlabel("t", loc="right")
ax.set_ylabel("p", loc="top")
```
